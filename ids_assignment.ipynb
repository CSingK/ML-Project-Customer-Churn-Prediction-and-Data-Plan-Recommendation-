```python
unique_seniorcitizen=df['Senior Citizen'].unique()
print("senior citizen",unique_seniorcitizen)

unique_dependent=df['Dependents'].unique()
print("dependent",unique_dependent)

unique_phoneservice=df['Phone Service'].unique()
print("phone service",unique_phoneservice)

unique_multiplelines=df['Multiple Lines'].unique()
print("multiple lines",unique_multiplelines)

unique_onlinesecurity=df['Online Security'].unique()
print("online security",unique_onlinesecurity)

unique_onlinebackup=df['Online Backup'].unique()
print("online backup",unique_onlinebackup)

unique_deviceprotection=df['Device Protection'].unique()
print("device protection",unique_deviceprotection)

unique_techsupport=df['Tech Support'].unique()
print("techsupport",unique_techsupport)

unique_streamingtv=df['Streaming TV'].unique()
print("streaming TV", unique_streamingtv)

unique_streamingmovies=df['Streaming Movies'].unique()
print("streaming movies", unique_streamingmovies)

unique_churn=df['Churn'].unique()
print("churn",unique_churn)

```
